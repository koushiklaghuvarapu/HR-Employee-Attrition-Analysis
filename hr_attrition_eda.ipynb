```python
# Q9 — Attrition by Age group
df['AgeGroup'] = pd.cut(df['Age'],
    bins=[18, 25, 35, 45, 60],
    labels=['18-25', '26-35', '36-45', '46-60'])
```

```python
y=df.groupby(['AgeGroup','Attrition']).agg({'Attrition':['count']})
y.reset_index(inplace=True)
y.columns=['AgeGroup','Attrition','Attrition_count']
pivot1=y.pivot(index='AgeGroup',columns='Attrition',values='Attrition_count')
pivot1['Attrition_Rate']=(pivot1['Yes']/(pivot1['Yes']+pivot1['No']))*100
pivot1
```

```python
sns.barplot(x='AgeGroup', y='Attrition_Rate', data=pivot1)
plt.title('Attrition Rate by Age Group (%)')
plt.show()
```

```python
# Q10 — Attrition by OverTime
overtime_attr = df.groupby('OverTime')['Attrition']\
    .apply(lambda x: (x=='Yes').sum() / len(x) * 100)\
    .reset_index()
overtime_attr.columns = ['OverTime', 'Attrition_Rate']
print(overtime_attr)

sns.barplot(x='OverTime', y='Attrition_Rate', data=overtime_attr)
plt.title('Attrition Rate by OverTime (%)')
plt.show()
```

```python
# Q11 — Salary distribution by Attrition
sns.boxplot(x='Attrition', y='MonthlyIncome', data=df)
plt.title('Monthly Income vs Attrition')
plt.show()
```

```python
# Average salary comparison
print(df.groupby('Attrition')['MonthlyIncome'].mean())
```

```python
# Q12 — Attrition by JobRole
jobrole_attr = df.groupby('JobRole')['Attrition']\
    .apply(lambda x: (x=='Yes').sum() / len(x) * 100)\
    .reset_index()
jobrole_attr.columns = ['JobRole', 'Attrition_Rate']
jobrole_attr = jobrole_attr.sort_values('Attrition_Rate', ascending=False)
```

```python
sns.barplot(y='JobRole', x='Attrition_Rate', data=jobrole_attr)
plt.title('Attrition Rate by Job Role (%)')
plt.show()
```

```python

# Q13 — WorkLifeBalance vs Attrition
wlb_attr = df.groupby('WorkLifeBalance')['Attrition']\
    .apply(lambda x: (x=='Yes').sum() / len(x) * 100)\
    .reset_index()
wlb_attr.columns = ['WorkLifeBalance', 'Attrition_Rate']
wlb_attr
```

```python
sns.barplot(x='WorkLifeBalance', y='Attrition_Rate', data=wlb_attr)
plt.title('Attrition Rate by Work Life Balance (1=Bad, 4=Best)')
plt.show()
```

```python
# Q14 — Years at company vs Attrition
sns.boxplot(x='Attrition', y='YearsAtCompany', data=df)
plt.title('Years at Company vs Attrition')
plt.show()
```

```python
print(df.groupby('Attrition')['YearsAtCompany'].mean())
```

```python
# Q15 — Correlation heatmap (numeric columns only)
numeric_cols = df.select_dtypes(include=[np.number]).columns
corr_matrix = df[numeric_cols].corr()

plt.figure(figsize=(14, 10))
sns.heatmap(corr_matrix, annot=True, fmt='.2f',
            cmap='coolwarm', center=0)
plt.title('Correlation Matrix')
plt.show()
```

```python

# Q16 — Attrition by MaritalStatus
marital_attr = df.groupby('MaritalStatus')['Attrition']\
    .apply(lambda x: (x=='Yes').sum() / len(x) * 100)\
    .reset_index()
marital_attr.columns = ['MaritalStatus', 'Attrition_Rate']

sns.barplot(x='MaritalStatus', y='Attrition_Rate', data=marital_attr)
plt.title('Attrition Rate by Marital Status (%)')
plt.show()
```

```python
# Q17 — JobSatisfaction vs Attrition
jobsat_attr = df.groupby('JobSatisfaction')['Attrition']\
    .apply(lambda x: (x=='Yes').sum() / len(x) * 100)\
    .reset_index()
jobsat_attr.columns = ['JobSatisfaction', 'Attrition_Rate']

sns.barplot(x='JobSatisfaction', y='Attrition_Rate', data=jobsat_attr)
plt.title('Attrition Rate by Job Satisfaction (1=Low, 4=High)')
plt.show()
```

```python
# Q18 — Distance from home vs Attrition
sns.boxplot(x='Attrition', y='DistanceFromHome', data=df)
plt.title('Distance From Home vs Attrition')
plt.show()
```

```python
# Q19 — Multiple factors combined
# OverTime + JobLevel + Attrition
pivot2 = df.pivot_table(
    values='Attrition',
    index='JobLevel',
    columns='OverTime',
    aggfunc=lambda x: (x=='Yes').sum() / len(x) * 100
)
pivot2
```

```python
sns.heatmap(pivot2, annot=True, fmt='.1f', cmap='Reds')
plt.title('Attrition Rate % by JobLevel and OverTime')
plt.show()
```
